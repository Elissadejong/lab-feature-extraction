## 2. Plot all the categorical variables with the proper plot. What can you see?

````python
a = 7  # number of rows
b = 2  # number of columns
c = 1  # initialize plot counter

fig = plt.figure(figsize=(15,35))

for i in categoricals:
    plt.subplot(a, b, c)
    plt.title('{}, subplot: {}{}{}'.format(i, a, b, c))
    plt.xlabel(i)
    sns.countplot(categoricals[i])
    c = c + 1

plt.show()
````

# What can we learn from these plots?
The variable 'response' is enormously imbalanced and may be redundant.

The values 'master' & 'doctor' in the variable 'education' could be bucketed into 1 value.

The variable 'employmentstatus' could be bucketed into just two categories: employed/unemployed.

The variable 'marital_status' could be bucketed into just two categories: married/single.

Other variables are also very imbalanced so, before we apply engineering techniques to possibly strengthen our model, we need to check if and/or how these variables are initially correlated with each other.
